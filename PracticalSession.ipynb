## Reading DataSet

````python
df = pd.read_csv("iris.csv", sep = ",")
````

````python
X = df[['SepalLengthCm','SepalWidthCm', 'PetalLengthCm','PetalWidthCm']]
````

````python
df['Species'].unique()
````

````python
X.plot(kind = 'box')
````

````python
X.plot(kind = 'hist')
````

````python
plt.figure(figsize=(10,10))
tags = X.keys()
counter = 1
for i in tags:
    plt.subplot(2,2,counter)
    counter = counter + 1
    X[i].plot(kind = 'hist')
    plt.title(i)
````

````python
X.corr()
````

````python
sns.heatmap(X.corr(), annot= True)
````

````python
sns.pairplot(X)
````

````python
g = sns.FacetGrid(data = df, col = 'Species')
g = g.map(sns.distplot, "SepalLengthCm")
````
